## Realmente necesitamos aprender otro lenguaje?

El problema de los dos lenguajes.

En primer lugar, al elegir un lenguaje se considera la facilidad de aprendizaje, la generaci칩n de c칩digo r치pidamente, esto es la **productividad** (_productive_ o _high-level languages_) que ofrece el lenguaje: python, ruby, matlab
En segundo lugar, en diversos problemas a tratar es indispensable considerar  el **rendimiento** del lenguaje de programaci칩n (_performance_ o _lower-level languages_) y la generaci칩n del c칩digo m치quina eficiente: Fortran, C, C++. En general un lenguaje es atractivo si es productivo, sin embargo su rendimiento no es 칩ptimo y viceversa; por tanto, la elecci칩n de productividad o rendimiento depender치 de las necesidades particulares del problema a resolver.

Adicional a lo anterior la **generalidad** del lenguaje, esto es tomar en cuenta si se trata de un lenguaje de prop칩sito genenal (_general purpose language_) o de dominio espec칤fico (_domain-specific language_). Es posible generar c칩digo eficiente y r치pido de manera sencilla si se trabaja cn lenguajes de dominio espec칤fico o con librer칤as especiales sin embargo el alcance (_scope_) de dichas herramientas ser치 un tanto limitado.

As칤, para lograr la productividad, rendimiento y generalidad ( _Productivity vs Performance vs Generality_) es necesario utilizar dos lenguajes. El prototipado de las soluciones se realiza por medio de un lenguaje de alto nivel como Python o MATLAB lo que proporciona una prueba de concepto r치pidamente; una vez hecho lo anterior, se realiza la traducci칩n del protototipo a otro lenguaje como C o C++ para realizar los c치lculos de manera eficiente.

Evitar el uso de dos lenguajes de programaci칩n es la raz칩n primordial de Julia.

El siguiente c칩digo muestra la jerarqu칤a de tipos de datos de Julia

````julia
function _show_subtype_tree(mytype, printlevel)
    allsubtypes = subtypes(mytype)
    for cursubtype in allsubtypes
        print("\t"^printlevel)
        println("|___",cursubtype)
        printlevel += 1
        _show_subtype_tree(cursubtype, printlevel)
        printlevel -= 1
    end
end
function show_type_tree(T)
    println(T)
    _show_subtype_tree(T,0)
end
show_type_tree(Number)
````

### Short-circuit 

El operador `&` hace la evaluaci칩n de dos expresiones booleanas por medio de la operaci칩n `AND`. El operador `&&` hace la misma evaluaci칩n pero en caso que la primera evaluaci칩n sea `false` no hace la segunda evaluaci칩n. 

````julia
# operador para AND
false & true
````

````julia
# short-circuit operator para AND
false && true
````

````julia
false && (println("Hola"); true)
````

````julia
true && (println("Hola"); false)
````

Por otra parte, el operador `&&` no requiere que la segunda expresi칩n devuelva el valor de `true` o `false`, veamos un ejemplo:

````julia
x = 5
(x > 0) && error("x no puede ser mayor que 0")
````

````julia
(x < 0) && (error("x no puede ser mayor que 0"); true)
````

De forma similar, el operador `||` es utilizado para hacer la evaluaci칩n de corto circuito de la operaci칩n `OR`. Dado un `true` en la primera expresi칩n no se eval칰a la segunda.

````julia
# operador para OR
true | false
````
